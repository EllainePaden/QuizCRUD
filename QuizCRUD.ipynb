```
import os

questions = []
choices = []
answers = []
players = []

#Input Name of players
def Introduction():
   Introduction.name = input('Kindly Enter your Name: ')
   print("Hello" + "\t" + Introduction.name.title() + "!" + "\nWelcome to the Planetary Quiz" + "\nChoose the Correct Answer to the following questions.")

#To save questions, choices and answers on a text file
def save_question(questions,choices,answers):
    with open("questions.txt","w") as file:
        for question in questions:
            file.write('%s\n' % question)
    with open("choices.txt","w") as file:
        for choice in choices:
            file.write('%s\n' % choice)
    with open("answers.txt","w") as file:
        for answer in answers:
            file.write('%s\n' % answer)

#Default questions, choices and answers    
def config_game():
    questions = ['Question','How many planets are there in the Solar System?',
              'What planet is closest to the sun?',
              'Which is the coldest planet of the Solar System?',
              'Which is the densest planet?',
              'Which planet is not a terrestial planet?',
              'What is the eight planet from the sun?',
              'Which planet has the largest ring system?',
              'What is the rotational period of Earth?',
              'Which is shorter, Mercury  day or year?',
              'Which planet has the hottest temperature?']
    choices = ['Choices','A. 3 \tB. 9','A. Mars \tB. Mercury','A. Pluto \tB. Saturn','A. Earth \tB. Jupiter',
               'A. Saturn \tB. Mars','A. Neptune \tB. Jupiter','A. Saturn \tB. Jupiter','A. 24 hours \tB. 36 hours','A. Day \tB. Year',
              'A. Venus \tB. Mars']
    answers = ['Answers','B','B','A','A','A','A','A','A','B','A']

    save_question(questions,choices,answers) 

#Putting file content into the list
def initialize_game():
    with open("questions.txt") as file:
        for question in file: 
            question = question.strip()
            questions.append(question)
    with open("choices.txt") as file:
        for choice in file: 
            choice = choice.strip()
            choices.append(choice)
    with open("answers.txt") as file:
        for answer in file: 
            answer = answer.strip()
            answers.append(answer)

#Preparing questions, choices and answers for the game
def load_game():
    if os.path.exists('questions.txt'):
        initialize_game()
    else:
        config_game()
        initialize_game()

#Display current questions, choices and answers
def display_game():
    num_q = len(questions) 
    i = 1

    print ("\n")
    while i < num_q:
      print(str(i) + ". " + questions[i])
      print(choices[i])
      print("Answer: {0}\n".format(answers[i]))
      i += 1

#Menu 1
def play_game():
    num_q = len(questions) 
    i = 1
    score=0

    while i < num_q:

      print(str(i) + ". " + questions[i])
      print(choices[i])
      ask = input("Choose A or B: ")

      if ask.upper() == answers[i]:
        print("You are Correct\n")
        score += 1
      else:
        print("Sorry, you got it wrong!\n")

      i += 1

    print("Your Score is {0}".format(score))

    #Record Players
    P1 = Player(Introduction.name,score)
    players.append(P1)
    save_player(players)



#Menu 2
def change_question():

    display_game()

    number=input("Enter question number to update or 0 to add new question: ")
    new_question=input("Enter New question: ")
    new_choices=input("Enter choices: ")
    new_answer=input("Enter Answer: ")

    if number == '0':
        questions.append(new_question)
        choices.append(new_choices)
        answers.append(new_answer)
    else:
        questions[int(number)] = new_question
        choices[int(number)] = new_choices
        answers[int(number)] = new_answer

    save_question(questions,choices,answers) 

    display_game()

    print("You have sucessfully update/add the game question.")

#Menu 3
def delete_question():

    display_game()

    rem = input("Enter question number to delete: ")

    i = int(rem)

    if i == 0:
        print("Invalid Question Number")
    else:
        questions.pop(i)
        choices.pop(i)
        answers.pop(i)
        print("Question {0} has been deleted".format(i))
    
    save_question(questions,choices,answers)
    display_game()

#Player Object
class Player ():
   def __init__(self,name,score):
        self.name = name
        self.score = score  

#Save Players into file
def save_player(players):
  with open("players.txt","w") as file:
        for player in players:
            file.write('%s\n' % "Player Name:{0} | Score:{1}\n".format(player.name,player.score))

#Game Start Here

load_game()
print('\n')

while True:
    
    game = input("\nWELCOME \n[1] Play Game \n[2] Update/Add Question \n[3] Delete Question \n[0] Exit \nSelect: ")


    if game == '1':
        os.system('cls')
        Introduction()
        play_game()
        continue
    elif game == '2':
        os.system('cls')
        change_question()
        continue
    elif game == '3':
        os.system('cls')
        delete_question()
        continue
    elif game == '0':
        break

#Display Game Result
with open("players.txt","r") as file1: 
  text = file1.read() 
  print("\nGame Results:")
  print(text)

print('\nGood Bye, Thank you for playing!')
```
